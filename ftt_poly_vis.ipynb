```python
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

%matplotlib widget
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
# set up plot
fig, ax = plt.subplots(figsize=(8, 8))
# ax.set_ylim([-4, 4])
ax.grid(True)
 
# generate x values
x = np.linspace(0, 0.5, 300)
 
 
def my_sine(x, a, b, c, d, e, f):
    """
    Return a sine for x with angular frequeny w and amplitude amp.
    """
    w = 2*np.pi*x
    poly = x
    return a*poly + b*np.sin(w) + c*np.cos(w) + d * poly**2 + e*np.sin(2*w) + f*np.cos(2*w) 
 
 
@widgets.interact(
    a=(-100, 100, 0.01), 
    b=(-100, 100, 0.01), 
    c=(-100, 100, 0.01), 
    d=(-100, 100, 0.01), 
    e=(-100, 100, 0.01), 
    f=(-100, 100, 0.01))
def update(a = 1.0, b=1., c=1., d=1., e=1., f=1.):
    """Remove old lines from plot and plot new one"""
    [l.remove() for l in ax.lines]
    ax.plot(x, my_sine(x, a, b, c, d, e, f), color='C0')
```
